```python
## view exp history use python2
## ==> view exp8

# file_name = 'for_test.txt' 
file_name = 'myout_DQN_try.txt'
data = None
with open(file_name, 'r') as f:
    data = f.read()

# print(data)
```

```python
step_list = []
length_list = []
area_list = []
reward_list = []

step_word = 'Step '
length_word = 'loop length = '
area_word = 'loop area = '
reward_word = ' reward: '


## cut before 'Game start'
data = data.split('Game start')[-1]
# print(data)


def get_valut(data, target_list, target_word):
    temp_data = data.split(target_word)
    for temp_step in temp_data[1:]:
        target_value = temp_step.split(' ')[0]
        if target_value[-1] == ',':
            target_value = target_value[:-1]
        target_value = float(target_value)
        target_list.append(target_value)
    return target_list


step_list = get_valut(data, step_list, step_word)
length_list = get_valut(data, length_list, length_word)
area_list = get_valut(data, area_list, area_word)
reward_list = get_valut(data, reward_list, reward_word)
# print('step_list: {} \nlength_list:{} \narea_list:{} \nreward_list:{}'.format(step_list, length_list, area_list, reward_list))
```

```python
## plot
import matplotlib.pyplot as plt
import numpy as np

def plot_fuc(target_list, title, y_lable, y_ticks=None, x_lim=(0, 2000), y_lim=None):
    plt.plot(target_list)
    plt.title(title)
    plt.ylabel(y_lable)
    plt.xlabel('episode')
    if y_ticks != None:
        plt.yticks(y_ticks)
    if y_lim != None:
        plt.ylim(y_lim)
    if x_lim != None:
        plt.xlim(x_lim)
    # plt.yticks(np.arange(0, 2.5, 0.25))
    # plt.legend(['train', 'test'], loc='upper right')
    # plt.savefig(save_path + pic_name_list[temp_index] + '_loss.png', dpi=300, format="png")
    plt.show()


plot_fuc(target_list=step_list, title='step_list', y_lable='step')
# plot_fuc(target_list=step_list, title='step_list', y_lable='step', y_ticks=np.arange(0, 300, 50))
plot_fuc(target_list=step_list, title='step_list', y_lable='step', y_lim=(0,300))


# plot_fuc(target_list=length_list, title='length_list', y_lable='length')
# plot_fuc(target_list=area_list, title='area_list', y_lable='area')

plot_fuc(target_list=reward_list, title='reward_list', y_lable='reward')


```
